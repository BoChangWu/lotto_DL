```python
from selenium import webdriver
from selenium.common.exceptions import TimeoutException
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By
from time import sleep
import json
import os

options = webdriver.ChromeOptions()
options.add_argument("--start-maximized")
options.add_argument("--incognito")
options.add_argument("--disable-popup-blocking")
driver = webdriver.Chrome(options = options)

lotto_nums=[]
lotto_year=108
lotto_month=8

def visit():
    driver.get('https://www.taiwanlottery.com.tw/lotto/DailyCash/history.aspx')

def parse():
    nums=[]
    for i in range(24):
        for j in range(5):
            num=driver.find_elements(By.CSS_SELECTOR,f"span#D539Control_history1_dlQuery_SNo{j+1}_{i}")
#             print(num[0].text)
            nums.append(num[0].text)
        lotto_nums.append(nums)
        nums=[]

        #       nums.append(num.text)

        
def search():
    for m in range(lotto_month):
        #選擇查詢方法
        method_tag = driver.find_element(By.CSS_SELECTOR,
                                         "input#D539Control_history1_radYM")
        method_tag.click()

        # 選擇年份
        year_select_opt = driver.find_element(By.CSS_SELECTOR,
                                             f"select#D539Control_history1_dropYear option[value='{str(lotto_year)}']")
        year_select_opt.click()

        # 選擇月份 select tag的選擇方法
        month_select_opt=driver.find_element(By.CSS_SELECTOR,
                                             f"select#D539Control_history1_dropMonth option[value='{str(m+1)}']")
        month_select_opt.click()

        # 查詢
        search_btn=driver.find_element(By.CSS_SELECTOR,
                                      "input#D539Control_history1_btnSubmit")
        search_btn.click()

        parse()
        
    print(lotto_nums)
    print('='*30)


def saveJson():
    fp = open("539108.json","w",encoding='utf-8')
    fp.write(json.dumps(lotto_nums,ensure_ascii = False))
    fp.close()
if __name__ == '__main__':
    visit()
    search()
    saveJson()
    
    
```
